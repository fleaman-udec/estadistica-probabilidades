# Gráficos unidimensionales

## Histograma
- Es una representación aproximada de la distribución de una variable estadística.
- Se construye dividiendo el rango de valores en intervalos del mismo tamaño y contando cuántos de los datos caen en cada intervalo.
- Cada intervalo se representa mediante una barra cuya superficie (o altura) es proporcional a la frecuencia con que la variable está en ese intervalo.
- Se puede expresar en frecuencia absoluta o relativa al total de datos.

<center>
    <img src="figures/histogram-1.png" width="400"/>
</center>

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(16)
x1 = np.random.rand(100)*100
plt.hist(x1, bins=10, edgecolor='black', density=False)
plt.xlabel('intervalos'), plt.ylabel('variable estadística x')
plt.show()
```

## Gráfico de torta
- Tiene forma circular y se utiliza para representar porcentajes o proporciones.
- El arco (y ángulo) de cada porción es proporcional a la cantidad que representa.

<center>
    <img src="figures/pie-1.png" width="400"/>
</center>

```python
estaciones = ['verano', 'otoño', 'invierno', 'primavera']
cantidades = [15, 10, 5, 10]
plt.pie(x=cantidades, labels=estaciones, autopct='%1.1f%%')
plt.title('Preferencias de estaciones del año de 30 estudiantes')
plt.show()
```

## Gráfico de caja
- Se compone de una caja, bigotes y círculos, y busca resumidamente mostrar la distribución de una variable estadística.
- La caja sirve para visualizar la tendencia central y asimetría de la muestra por medio de sus cuartiles.
- Además, los bigotes y círculos sirven para visualizar la variabilidad fuera del primer y tercer cuartil.

<center>
    <img src="figures/boxplot-1.png" width="800"/>
</center>

```python
np.random.seed(46)
x2 = 50 + 10*np.random.randn(100)
box = plt.boxplot([x1, x2], labels=['caso 1', ' caso 2'])
#box['boxes'][1].set_facecolor('pink')
plt.ylabel('variable estadística x')
plt.show()
```

```python
fig, ax = plt.subplots()

bplot = ax.boxplot([x1, x2], notch=True, patch_artist=True, labels=['caso 1', 'caso 2'])

colors = ['lightblue', 'lightgreen']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)        

ax.yaxis.grid(True)
ax.set_ylabel('variable estadística x')
plt.show()
```

## Gráfico de violín
Es parecido al gráfico de caja, pero también muestra una estimación de la densidad de probabilidad de los datos (concepto a ver más adelante).

<center>
    <img src="figures/violin-1.png" width="800"/>
</center>

```python
box = plt.violinplot([x1, x2])
plt.ylabel('variable estadística x')
plt.show()
```

# Gráficos bidimensionales

## Gráfico de línea
- Es el gráfico que simplemente conecta con una línea los valores correspondientes de dos variables estadísticas en un plano Cartesiano.
- Comúnmente se usa cuando la variable del eje horizontal sigue alguna progresión ordenada (e.g. cronológica) o cuando representa datos categóricos en vez de numéricos.

<center>
    <img src="figures/lineplot-2.jpg" width="800"/>
</center>

```python
x = np.arange(0, 2, 0.01)
y = np.cos(2*np.pi*1.*x + 0.)

plt.plot(x, y)
plt.xlabel('tiempo [s]'), plt.ylabel('aceleración [mm/s]')
plt.grid()
plt.show()
```

```python
y1 = 5 + 2*x + np.sin(2*x) + np.random.rand(len(x))*0.1
y2 = 4 + 2*x + np.sin(2*x) + np.random.rand(len(x))*0.1

fig, ax = plt.subplots()
ax.plot(x, y1, label='dia')
ax.plot(x, y2, label='noche')
ax.set_xlabel('tiempo [h]'), ax.set_ylabel('temperatura [°C]')
plt.legend()
plt.show()
```

## Gráfico de línea con margen
Es una variación del gráfico de línea que incorpora la variabilidad de la variable en el eje vertical

```python
fig, ax = plt.subplots()
ax.fill_between(x, y1, y2, alpha=.5, linewidth=0)
ax.plot(x, (y1 + y2)/2, linewidth=2)
ax.set_xlabel('tiempo [s]'), ax.set_ylabel('temperatura [K]')
plt.show()
```

## Diagrama de dispersión 2D
- Los valores correspondientes de dos variables estadísticas se muestran como puntos en un plano.
- Se usa en el análisis exploratorio de datos para visualizar , comparar y relacionar valores numéricos.
- Es posible agregar colores o tamaños a los puntos para evaluar una dimensión adicional.

<center>
    <img src="figures/scatter-1.png" width="800"/>
</center>

```python
np.random.seed(88)
x3 = 50 + 10*np.random.randn(100)
x4 = 5.5 + 1.5*x3 - 3.3*np.log2(x3) + 15*np.random.rand(100)
x5 = -12.3*np.log2(x3) + 15*np.random.rand(100)

fig, ax = plt.subplots(ncols=2, figsize=(14, 6))
ax[0].scatter(x3, x4)
ax[0].set_xlabel('variable x3'), ax[0].set_ylabel('variable x4')
ax[1].scatter(x3, x5, color='red')
ax[1].set_xlabel('variable x3'), ax[1].set_ylabel('variable x5')

ax[0].grid(), ax[1].grid(), plt.tight_layout()
plt.show()
```

# Gráficos tridimensionales

## Mapa de calor
Muestra una variable estadística en un plano Cartesiano en función de otras dos variables usando un mapa de colores proporcional a su magnitud.

<center>
    <img src="figures/heatmap-1.png" width="800"/>
</center>

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

xv, yv = np.meshgrid(x, y)
zv = np.sin(np.sqrt(xv**2 + yv**2))

#cmaps = viridis, plasma, Reds
fig, ax = plt.subplots()
heatmap = ax.pcolormesh(xv, yv, zv, shading='auto', cmap='plasma')
ax.set_xlabel('variable x'), ax.set_ylabel('variable y')
cbar = plt.colorbar(heatmap)
cbar.set_label('variable z')
plt.show()
```

## Gráfico de superficie
- Es la extensión del gráfico de línea a tres dimensiones.
- Muestra mediante una superficie los valores de una variable estadística en función de otras dos variables que siguen una progresión ordenada.

<center>
    <img src="figures/surface-1.png" width="800"/>
</center>

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(xv, yv, zv, color='lightblue')
plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(xv, yv, zv, cmap='coolwarm', linewidth=0)
cbar = fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_zticks([])
plt.show()
```

## Diagrama de dispersión 3D
- Es la extensión del gráfico de dispersión a tres dimensiones.
- Muestra mediante puntos los valores correspondientes de tres variables estadísticas.
- También es posible darles distintos tamaños y colores a los puntos para agregar otra dimensión.
<center>
    <img src="figures/scatter3d-1.png" width="800"/>
</center>

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x3, x4, x5, marker='s', color='red', label='set 1')
ax.scatter(x1, x2, x3, marker='o', color='blue', label='set 2')
plt.legend()
plt.show()
```

## Gráfico de contorno
- Es una representación en un plano Cartesiano de una superficie tridimensional.
- Se consigue realizando "cortes" en la superficie a distintas alturas, los cuales se visualizan en el gráfico mediante un mapa de colores.

<center>
    <img src="figures/contour-1.png" width="800"/>
</center>

```python
fig, ax = plt.subplots()
cont = ax.contourf(xv, yv, zv, cmap='plasma', levels=5)
ax.set_xlabel('variable x'), ax.set_ylabel('variable y')
cbar = plt.colorbar(cont)
cbar.set_label('variable z')
plt.show()
```
